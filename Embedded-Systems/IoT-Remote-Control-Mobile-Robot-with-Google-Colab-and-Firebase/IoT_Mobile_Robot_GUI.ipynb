```python
import firebase_admin as fb
import ipywidgets as widgets
from firebase_admin import db, credentials
from IPython.display import clear_output, display

#firebase authentication
cred = credentials.Certificate("ESP32_Integration.json") #file JSON dari firebase, HARUS DI UPLOAD TERLEBIH DAHULU
if not fb._apps:
  fb.initialize_app(cred, {"databaseURL": "https://esp32-integration-test-default-rtdb.asia-southeast1.firebasedatabase.app/"}) #Link dari realtime database

# creating reference to root node
ref = db.reference("/")

#Untuk menampilkan isi data di firebase
ref.get()

#Hanya menampilkan data tertentu
db.reference("/OUTPUT/speed").get()


left_button = widgets.Button(description="<", button_style='success',layout = widgets.Layout(width='5%'))
right_button = widgets.Button(description=">", button_style='success',layout = widgets.Layout(width='5%'))
lurus_button = widgets.Button(description="o", button_style='warning',layout = widgets.Layout(width='5%'))
drive_button = widgets.Button(description='^', button_style='success',layout = widgets.Layout(width='16%'))
reverse_button = widgets.Button(description='v', button_style='success', layout = widgets.Layout(width='16%'))

print('SILAHKAN TEKAN')
ref.get()

def left_button_clicked(b):
  value = 3
  db.reference("/OUTPUT/command").set(value)

def right_button_clicked(b):
  value = 2
  db.reference("/OUTPUT/command").set(value)

def lurus_button_clicked(b):
  value = 0
  db.reference("/OUTPUT/command").set(value)

def drive_button_clicked(b):
  drive = 1
  db.reference("/OUTPUT/command").set(drive)

def reverse_button_clicked(b):
  value = 4
  db.reference("/OUTPUT/command").set(value)

def speed(s):
  speedcar = s
  db.reference("/OUTPUT/speed").set(speedcar)

menyamping = widgets.HBox([left_button,lurus_button, right_button])
display(widgets.VBox([drive_button,menyamping,reverse_button]))

widgets.interact(speed, s=(0,255,1))

left_button.on_click(left_button_clicked)
right_button.on_click(right_button_clicked)
lurus_button.on_click(lurus_button_clicked)
drive_button.on_click(drive_button_clicked)
reverse_button.on_click(reverse_button_clicked)
```
